```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from apyori import apriori
```

```python
data=pd.read_csv("groceries - groceries.csv")
data
```

```python
data.shape
```

```python
transactions=[]
for i in range(0, 5000):
    transactions.append([str(data.values[i,u]) for u in range(1, 32)])
```

```python
for sets in transactions:
    print(sets)
```

```python
association_rules=apriori(transactions,min_support=0.5,min_confidence=0.7,min_length=2)
```

```python
association_results=list(association_rules)
```

```python
print(len(association_results))
```

```python
association_results[0]
```

```python
for item in association_results:
    pair = item[0]
    items = [x for x in pair]
    print("Rule :"+ str(item[0]) + "->" + str(item[1]))
    print("Support : {}".format(item[1]))
    print("Confidence : {}".format(item[2][0][2]))
    print("Lift : {}".format(item[2][0][3]))
    print("\n------------------------------------------------\n")
```
